```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import pyarrow
```

```python
def read_parquet (folder_path, info) :
    files = os.listdir(folder_path)
    Specified_files = [file for file in files if info in file]
    dfs = [pd.read_parquet(os.path.join(folder_path, f))
           for f in Specified_files]
    df = pd.concat(dfs)
    return df
```

Q4:What is the longest match recorded in terms of duration?

```python
matches_time = read_parquet('data\\raw_match_parquet', 'time').reset_index()
matches_time = matches_time.drop('index', axis=1)
matches_time
```

```python
matches_periods = ['period_1', 'period_2', 'period_3', 'period_4', 'period_5']
for period in matches_periods :
    if not 'None' :
        matches_time[period] = matches_time[period].astype(int)
matches_time
```

```python
matches_time.fillna(0, inplace=True)
matches_time
```

```python
matches_periods = ['period_1', 'period_2', 'period_3', 'period_4', 'period_5']
for period in matches_periods:
    matches_time[period] = matches_time[period]/60
matches_time['match_duration'] = matches_time['period_1'] + matches_time['period_2'] + matches_time['period_3'] + matches_time['period_4'] + matches_time['period_5']
matches_time['match_duration'].max()
```

The longest match duration in this tennis tournoment is almost 213 minutes or 6 hours and 33 minutes

Q12:What is the average number of games per set in men's matches compared to women's matches? 

```python
power = read_parquet('data\\raw_tennis_power_parquet', 'power')
power
```

```python
games_per_set = power.merge(home_team, on = 'match_id', how = 'left')
games_per_set
```

```python
female_games_per_set = games_per_set[games_per_set['gender'] == 'F']
male_games_per_set = games_per_set[games_per_set['gender'] == 'M']
F_count = female_games_per_set.groupby('match_id', as_index = False)['set_num'].value_counts()
F_count
```

```python
Fmean_games_per_set = F_count['count'].mean()
Fmean_games_per_set
```

```python
M_count = male_games_per_set.groupby('match_id', as_index=False)['set_num'].value_counts()
M_count
```

```python
Mmean_games_per_set = M_count['count'].mean()
Mmean_games_per_set
```

Avarage of games per set in female competitions is 8 games and in male competitions is 9
